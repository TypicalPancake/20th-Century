```python
# Import country relationship file
import pandas as pd

df = pd.read_csv('country_relationships_20th.csv')
df.head()
```

```python
# Create/Import NewtorkX graph data frame
import networkx as nx

G = nx.from_pandas_edgelist(df, 'Country1', 'Country2')
```

```python
# Import/Create Static Graph with NetworkX
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=800)
plt.title("20th-Century Country Relationships (Static Network)")
plt.show()
```

```python
# Create interactive graph with Pyvis
from pyvis.network import Network

net = Network(notebook=True)
net.from_nx(G)
net.show("static_20th_century_network.html")
```

```python
# Apply Leiden Algorithm
from cdlib import algorithms

leiden_communities = algorithms.leiden(G)
print(leiden_communities.communities[:3])
```

```python
# Visualize Communities In Pyvis
com_net = Network(notebook=True)
com_net.from_nx(G)

colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown']

for i, community in enumerate(leiden_communities.communities):
    for node in community:
        com_net.get_node(node)['color'] = colors[i % len(colors)]

com_net.show("leiden_communities_network.html")
```

# The Leiden algorithm grouped countries into communities based on how closely they were connected, and the resulting partitions align well with major 20th-century alliances. For example, some clusters reflect World War II alliances, Cold War blocs, or regional coalitions. These groupings highlight the structured nature of international relations and reveal how certain countries were central or acted as bridges between different communities. Overall, the partition makes historical sense and supports the idea that geopolitical influence shaped global connections throughout the century.

```python
# Centrality Analysis
import matplotlib.pyplot as plt

# Calculate centrality measures
degree = nx.degree_centrality(G)
closeness = nx.closeness_centrality(G)
betweenness = nx.betweenness_centrality(G)

# Convert to sorted lists of (country, value)
sorted_degree = sorted(degree.items(), key=lambda x: x[1], reverse=True)
sorted_closeness = sorted(closeness.items(), key=lambda x: x[1], reverse=True)
sorted_betweenness = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)

# Function to plot a bar chart
def plot_centrality(data, title):
    labels, values = zip(*data)
    plt.figure(figsize=(12, 6))
    plt.bar(labels, values)
    plt.xticks(rotation=45, ha='right')
    plt.title(title)
    plt.ylabel("Centrality Score")
    plt.tight_layout()
    plt.show()

# Plot all three
plot_centrality(sorted_degree, "Degree Centrality of Countries")
plot_centrality(sorted_closeness, "Closeness Centrality of Countries")
plot_centrality(sorted_betweenness, "Betweenness Centrality of Countries")
```

# 1. Degree centrality measures the number of direct relationships a country has in the network. Countries with high degree centrality are often global powers or those involved in multiple conflicts or alliances.

# For example, if the United States, Germany, or United Kingdom have high degree scores, this reflects their extensive involvement in major 20th-century events like both World Wars, the Cold War, and various diplomatic treaties. High degree could also reflect geopolitical influence or colonial relationships.

# Countries with lower degree centrality may have remained neutral, had regional rather than global influence, or became significant only later in the century.
